```python
import sympy as sp
from sympy.plotting import plot
import math as  m
import numpy as n
```

```python
x = sp.Symbol('x')
y = -12*x**4*sp.sin(sp.cos(x))-18*x**3+5*x**2 + 10*x - 30

yprime = y.diff(x)
print(yprime)
```

```python
plot(y,(x, -100, 100))
```

```python
xx = sp.solve(yprime,x)
```

```python
print("Корни: ",xx)
```
